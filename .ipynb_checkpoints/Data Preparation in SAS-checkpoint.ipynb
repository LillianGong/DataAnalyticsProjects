# Data Preparation Project in SAS

The dataset is a 1994 Census Bureau Database of the United States of America https://www.census.gov/en.html. The dataset lists 15 variables used to determinate the income of the American citizen. 

### 1. Import data

```sas
* Import audit.xls and setup dataset Income; 
libname GROUP '/folders/myfolders/DataPre';

proc import datafile = '/folders/myfolders/DataPre/adult.xls'
            out=group.income
            DMBS=xls
            replace;
run;

data group.income;
     set group.income;
     ID=_n_;
run;

title 'List of dataset Income';
proc print data=group.income(obs=10);
run;
```

### 2. Check categorical variables.

#### 2.1 Check and correct errors in categorical variables.

```sas
* Show the frequency table;
title 'Frequency table for categorical variables';
proc freq data=group.income;
     table _character_ ;
run;
```

```sas
*Check errors for categorical variables - native_country;

*create country dataset with stanard country names;

proc import datafile='/folders/myfolders/DataPre/country.xlsx'
            out=group.country
            DBMS=xlsx
            replace;
run;

proc freq data=group.income noprint;
     table native_country / out=group.native_country nopercent nocum;
run;

*Merge two datasets Native_country and Country by Fuzzy marching methods;
proc sort data=group.native_country(rename=(native_country=country)) out=group.native_country1;
     by country;
run;

proc sort data=group.country(rename=(country_name=country)) out=group.country1;
     by country;
run;
     
proc sql;
   create table group.left_join as
   select *
   from group.native_country1 left join group.country1
   on  (compged(native_country1.country, country1.country) le 50);
quit ;

proc sql;
   create table group.income_country_check as
   select distinct * from group.left_join;
quit;

proc format;
   value $country_check '?'='Missing'
                        ' '='Error'
                      other='Valid';
run;

title 'Checking Missing values and Errors for variable native_country';
data _null_; 
    set group.income_country_check; 
    file print; 
    if put(country, $country_check.) = 'Missing' then 
    put 
    count " Obeservations are missing value for native_country "  ; 
    else if put(Check, $country_check.) = 'Error' then 
    put "Native_country-" country "has error value"; 
run; 
```


Errors in categorical variables - Native_country: 

- Value "Columbia" has error. It should be "Colombia".
- Value "England" has error. It should be "United Kindom".
- Value "Holand-Netherlands" has error. It should be "Netherlands".
- Value "Hong" has error. It is not independt country. It should be "China".
- Value "Iran" can not be seen as an error. Because it can be recognized as "Iran,Islamic Republic of".
- Value "Laos" can not be seen as an error. Because it can be recognized as "Lao People's Democratic Republic".
- Value "Outlying-US(Guam-USVI-etc)" has error. It should be "Guam and Virgin Islands, U.S.".
- Value "Scotland" has error. It is not independent country. It should be changed to "United Kingdom".
- Value "South" has error. It should be "South Africa".
- Value "Trinadad&Tobago" has error. It should be "Trinadad and Tobago".
- Value "Vietnam" has error. It should be "Viet Nam".
- Value "Yugoslavia" has error. It was no longer a country. It can be changed to "Former Yugoslavia".


```sas
* Modify the errors in categorical variables - Native_country;
data group.income_modify_nov24;
     set group.income;
     if native_country = "Columbia"  then native_country ="Colombia";
     else if native_country ="England" then native_country ="United Kindom";
     else if native_country ="Holand-Netherlands" then native_country = "Netherlands";
     else if native_country ="Hong" then native_country ="China";
     else if native_country ="Outlying-US(Guam-USVI-etc)" then native_country ="Guam and Virgin Islands, U.S.";
     else if native_country ="Scotland"  then native_country ="United Kindom";
     else if native_country ="South" then native_country ="South Africa";
     else if native_country ="Trinadad&Tobago" then native_country ="Trinadad and Tobago";
     else if native_country ="Vietnam" then native_country ="Viet Nam";
     else if native_country ="Yugoslavia" then native_country ="Former Yugoslavia";
run;

title 'Frequency table for modified variable - native_country';
proc freq data=group.income_modify_nov24;
     table native_country / nopercent nocum ;
run;
```

#### 2.2 Check and replace missing values for categorical variables.

```sas
* Check missing values.

proc format;
     value $cate_check '?'='Missing values'
                    other='Not missing values';
run;

title 'Checking missing values for categorical variables';
proc freq data=group.income;
     tables _character_ / nopercent nocum ;
     format _character_ $cate_check.;
run;
```

Missing values in categorical variables:
- Workclass: 1836 observations are missing for variable workclass.
- Occupation: 1843 observations are missing for variable occupation.
- Native_country: 583 observations are missing for variable native_country.

```sas
* Replace the missing values with the mode;

data group.income_modify_nov25;
    set group.income_modify_nov24;
    if occupation = '?' then occupation ='Prof-specialty';
    if workclass='?' then workclass='Private';
    if native_country = '?' then native_country='United-States';
run;

title 'Frequency table for categorical variables after repalcing missing values';
proc freq data=group.income_modify_nov25; 
    tables _character_ / nocum nopercent ; 
    format _character_ $cate_check.; 
run; 
       
```

#### 2.3 Group values of categorical variables.

```sas
* Group the variables by combining some smallest groups into several subgroups;


data group.income_modify_nov25_group;
     set group.income_modify_nov25;
     
*Group variable-workclass into Goverment,Private,Self-employed,Unemployed; 
     if workclass in ('Federal-gov','Local-gov' ,'State-gov') then workclass='Government';
     else if workclass in ('Self-emp-inc','Self-emp-not-inc') then workclass='Self-employed';
     else if workclass in ('Never-worked','Without-pay') then workclass='Unemployed';
     
*Group variable-education into Before HS,HS-grad,After HS;  
     if education in ('Preschool','1st-4th','5th-6th','7th-8th','9th','10th','11th','12th') then education='Before HS';
     else if education in ('Assoc-acdm','Assoc-voc','Bachelors','Doctorate','Masters','Prof-school','Some-college') 
          then education='After HS';
          
*Group variable-marital_status into Married, Not-married, Never-married;
     if marital_status in ('Divorced','Separated','Widowed') then marital_status='Not-married';
     else if marital_status in ('Married-AF-spouse','Married-civ-spouse','Married-spouse-absent') then marital_status='Married';
     
*Group variable-native_country into North America, South America, Asia, Europe, Africa;
     if native_country in ('China','India','Iran','Cambodia','Japan', 'Laos','Philippines' ,'Vietnam' ,'Taiwan', 'Thailand')
        then native_country='Asia';
     else if native_country in ('Canada','United-States','Puerto-Rico','Mexico') then native_country='North America';
     else if native_country in ('United Kindom','France', 'Germany' ,'Greece','Netherlands','Hungary','Ireland','Italy',
     'Poland','Portugal','Former Yugoslavia') then native_country='Europe';
     else if native_country in('Colombia','Cuba','Dominican-Republic','Ecuador','El-Salvador','Guatemala','Haiti','Honduras',
                            'Nicaragua','Guam and Virgin Islands, U.S.','Peru', 'Jamaica','Trinadad&Tobago') 
      then native_country='Latin and South America';
      else native_country='Africa';
      
*Group variable-occupation into management, profession, sales, service, labor, Admin, military;
      if occupation in ('Craft-repair','Farming-fishing','Handlers-cleaners','Machine-op-inspct','Transport-moving') 
        then occupation='Labor';
      else if occupation in ('Other-service','Priv-house-serv','Protective-serv') then occupation='Service';
      else if occupation in ('Tech-support','Prof-specialty') then occupation='Profession';
run;

title 'Frequency tables for categotical variables after grouping';
proc freq data=group.income_modify_nov25_group;
     table _character_ / nopercent nocum ;
run;       
            
```

### 3. Check numeric variables

#### 3.1 Check and impute missing values

```sas
* check the descriptive stats;
proc univariate data = group.income_modify_non25_dummy;
var Age fnlwgt education_num capital_gain capital_loss hours_per_week;
run;
```

'99999' in capital_gain is supposed to be missing value, which is suggested to be imputed with median.

```sas
* calculate new median/mean after deleting missing values in 'capital_gain';
data group.income_modify_capgain;
   set group.income_modify_non25_dummy;
   if capital_gain = 99999 then delete;
run;

title 'New stats for capital_gain after deleting missing values';
proc univariate data =group.income_modify_capgain;
    var capital_gain;
run;
```

Median is different with Mean. In this situation, the missing value should be imputed with the Median, which is '0'.

```sas
* impute with the Median ('0') on original dataset which has all missing values '99999';
data group.income_modify_nov26;
   set group.income_modify_non25_dummy;
   if capital_gain = 99999 then capital_gain = 0;
run;

title 'New stats for capital_gain after immupatition';
proc univariate data =group.income_modify_nov26;
    var capital_gain;
run;
```

#### 3.2 Check outliers

```sas
*Detect outliers on numeric variable - fnlwgt by using IQR, which ignore the distribution of variables; 
*To increase the records for following models, Relax the definition of outliers to 3*IQR.

title "Outliers Based on Interquartile Range";
proc means data = group.income_modify_nov26 noprint;
   var fnlwgt;
   output out=Tmp1
          Q1=
          Q3=
          QRange= / autoname;
run;

title 'IQR for variable fnlwgt';
proc print data=Tmp1; 
run;


title 'Outliers for variable fnlwgt';
data _null_;
   file print;
   set group.income_modify_nov26(keep=ID fnlwgt);
   if _n_ = 1 then set Tmp1;
   if fnlwgt le fnlwgt_Q1 - 3*fnlwgt_QRange or
      fnlwgt ge fnlwgt_Q3 + 3*fnlwgt_QRange then
      put "Possible Outlier for ID " ID "fnlwgt is " fnlwgt;
run;
```

```sas
*Detect outliers on numeric variable - education_num by using IQR, which ignore the distribution of variables; 
*To increase the records for following models, Relax the definition of outliers to 3*IQR.

title "Outliers Based on Interquartile Range";
proc means data = group.income_modify_nov26 noprint;
   var education_num;
   output out=Tmp2
          Q1=
          Q3=
          QRange= / autoname;
run;

title 'IQR for variable education_num';
proc print data=Tmp2; 
run;


title 'Outliers for variable education_num';
data _null_;
   file print;
   set group.income_modify_nov26(keep=ID education_num);
   if _n_ = 1 then set Tmp2;
   if education_num le education_num_Q1 - 3*education_num_QRange or
      education_num ge education_num_Q3 + 3*education_num_QRange then
      put "Possible Outlier for ID " ID "education_num is " education_num;
run;
```

There are no outliers for variable -education_num by using definition of 3*IQR.

#### 3.3 Remove outliers
As numeric algorithms (linear regression, k-nearest neighbor, K-Means clustering, and Principal Component Analysis) are senstitive to outliers, while Decision tree is not. In terms of  variable fnlwgt, only 0.4% of which are outliers, we can seperate the possibles outliers into a new dataset as a subset for modeling in case the outliers might be the good predictor of the target variable.

```sas
*Remove outliers based on interquartile method and store them in new dataset called outliers;

title "Dataset after removing outliers for fnlwgt";

data group.income_no_fnlwgt_outliers;
   file print;
   set group.income_modify_nov26;
   if _n_ = 1 then set Tmp1;
   if fnlwgt le fnlwgt_Q1 - 3*fnlwgt_QRange and not missing(fnlwgt) or
      fnlwgt ge fnlwgt_Q3 + 3*fnlwgt_QRange then
      delete;
    drop fnlwgt_Q1 fnlwgt_Q3 fnlwgt_QRange _TYPE_ _FREQ_;
run;

data group.fnlwgt_outliers;
   file print;
   set group.income_modify_nov26;
   if _n_ = 1 then set Tmp1;
   if fnlwgt le fnlwgt_Q1 - 3*fnlwgt_QRange and not missing(fnlwgt) or
      fnlwgt ge fnlwgt_Q3 + 3*fnlwgt_QRange then
      output;
    drop fnlwgt_Q1 fnlwgt_Q3 fnlwgt_QRange _TYPE_ _FREQ_;
run;


title 'Stats for dataset without outliers of fnlwgt';
proc univariate data=group.income_no_outliers;
var fnlwgt;
run;

title 'Stats for outliers of fnlwgt';
proc univariate data=group.outliers;
var fnlwgt;
run;
```

After removing outliers, we get the new distribution of 'fglwgt' with less skewness which is smaller than 1.

#### 3.4 Create a dummy variable to indicate outliers
Variable 'hours_per_week'  has a large amount of outliers based on 3*IQR method. We can create a dummy variable to indicate outliers.

```sas
title 'Dummy variable to indicate outliers for hours_per_week variable';
proc means data = group.income_modify_nov26 noprint;
   var hours_per_week;
   output out=Tmp3
          Q1=
          Q3=
          QRange= / autoname;
run;

data group.income_hpw_outliers;
   file print;
   set group.income_no_fnlwgt_outliers;
   if _n_ = 1 then set Tmp3;
   if hours_per_week le hours_per_week_Q1 - 3*hours_per_week_QRange or
      hours_per_week ge hours_per_week_Q3 + 3*hours_per_week_QRange then
      hours_per_week_outlier = 1;
   else hours_per_week_outlier = 0;
   drop hours_per_week_Q1 hours_per_week_Q3 hours_per_week_QRange _TYPE_ _FREQ_;
run;

title 'Frequence for dummy variable hours_per_week_outlier';
proc freq data=group.income_hpw_outliers;
    tables hours_per_week_outlier;
run;
```

#### 3.5 Convert data to dummy variables

90% values of capital_gain and 95% of capital_loss are '0'. To leverage these two variables, we can convert them to dummy variables.

```sas
* convert non-zero value of capital_gain and capitao_loss into '1' with meaning of gain and loss respectively;

data group.income_numeric_dummy;
   set group.income_hpw_outliers;
   if capital_gain <> 0 then capital_gain = 1;
   if capital_loss <> 0 then capital_loss = 1;
run;

title 'List of dataset after converting to dummy variables';
proc print data= group.income_numeric_dummy(obs=10);
run;

title 'Frequence for dummy capital_gain and capital_loss';
proc freq data=group.income_numeric_dummy;
    tables capital_gain capital_loss;
run;

title 'Contents of dataset after cleaning numeric variables';
proc contents data = group.income_numeric_dummy;
run;
```

#### 3.6  Test normality and Data transformation

```sas

title "Histogram for fnlwgt variable";
proc sgplot data=group.income_numeric_dummy; 
histogram fnlwgt; 
run; 
```

```sas
*QQ Plot of variable fnlwgt;
title"fnlwgt Plots";

proc univariate data=group.income_numeric_dummy  NORMAL PLOT;
   var fnlwgt;
run;
```

```sas
* Create new dataset with a new variable called fnlwgt_root which contains the value of root transformation. It has a better performance than log tranformation;

Data group.income_fnlwgtroot; 
   SET group.income_numeric_dummy;
   fnlwgt_root= (fnlwgt)**0.5;
RUN;


title 'Stats for fnlwgt_root variable';

PROC UNIVARIATE DATA =group.income_fnlwgtroot NORMAL PLOT; 
   var fnlwgt_root;
RUN;
```

After applying log, skewness is now between -0.5 and 0.5, which means the variable fnlwgt_root is now noramlly distributed.
